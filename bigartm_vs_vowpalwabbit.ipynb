## BigARTM

To run the experiment for BigARTM we use script provided below. In this script

- $ns$ is the array of points on axis x where quality measures are estimated
- $Result$ is the matrix of quality measures on points from ns
- $collection$_$name$ is the name of collection in experiment
- $data$_$folder$ is the path to your collection dataset (files docword.collection_name.txt and vocab.collection_name.txt)
- $target$_$folder$ is the path to batches of collection


```python
import artm.messages_pb2, artm.library, sys

# Create list of quality measures
# Result[0] = Perplexity
# Result[1] = Phi_sparsity
# Result[2] = Theta_sparsity
# Result[3] = Average_kernel_size
# Result[4] = Average_kernel_purity
# Result[5] = Average_kernel_contrast
Result  = [[] for i in range(6)]
ns = [2, 5, 10, 15, 20, 25, 30, 35, 40, 45, 50]
ns_idx = 0

# Parse collection
data_folder = 'C:/local/bigartm/BigARTM_v0.5.5_x64/python/examples/'
target_folder = 'C:/local/bigartm/BigARTM_v0.5.5_x64/python/examples/nips'
collection_name = 'nips'
unique_tokens = artm.library.Library().ParseCollectionOrLoadDictionary(
  data_folder + 'docword.'+ collection_name + '.txt',
  data_folder + 'vocab.' + collection_name + '.txt',
  target_folder)

# Create local node controller
create_local_node_controller = True
if (create_local_node_controller):
  node_controller = artm.library.NodeController('tcp://*:5555')

# Create master component and infer topic model
proxy_config = artm.messages_pb2.MasterProxyConfig()
proxy_config.node_connect_endpoint = 'tcp://localhost:5555'
proxy_config.config.disk_path = target_folder


with artm.library.MasterComponent(proxy_config) as master:

  dictionary           = master.CreateDictionary(unique_tokens)
  
  perplexity_score     = master.CreatePerplexityScore()
  sparsity_theta_score = master.CreateSparsityThetaScore()
  sparsity_phi_score   = master.CreateSparsityPhiScore()
  topic_kernel_score   = master.CreateTopicKernelScore()
  top_tokens_score     = master.CreateTopTokensScore()

  # Configure basic regularizers
  #smooth_sparse_theta_reg  = master.CreateSmoothSparseThetaRegularizer()
  #smooth_sparse_phi_reg    = master.CreateSmoothSparsePhiRegularizer()
  decorrelator_reg     = master.CreateDecorrelatorPhiRegularizer()

  # Configure the model
  model = master.CreateModel(topics_count = 10, inner_iterations_count = 10)
  model.EnableScore(perplexity_score)
  model.EnableScore(sparsity_phi_score)
  model.EnableScore(sparsity_theta_score)
  model.EnableScore(topic_kernel_score)
  model.EnableScore(top_tokens_score)
  #model.EnableRegularizer(smooth_sparse_theta_reg, -0.1)
  #model.EnableRegularizer(smooth_sparse_phi_reg, -0.2)
  model.EnableRegularizer(decorrelator_reg, 500000)

  for iter in range (51):
    master.InvokeIteration(1)
    master.WaitIdle();
    model.Synchronize();
    
    if iter == ns[ns_idx]:
        ns_idx += 1
        Result[0].append(perplexity_score.GetValue(model).value)
        Result[1].append(sparsity_phi_score.GetValue(model).value)
        Result[2].append(sparsity_theta_score.GetValue(model).value)
        Result[3].append(topic_kernel_score.GetValue(model).average_kernel_size)
        Result[4].append(topic_kernel_score.GetValue(model).average_kernel_purity)
        Result[5].append(topic_kernel_score.GetValue(model).average_kernel_contrast)

  artm.library.Visualizers.PrintTopTokensScore(top_tokens_score.GetValue(model))

#Dispose local node controller
if (create_local_node_controller):
  node_controller.Dispose()
```

```python
fig, ax = plt.subplots(figsize = (8, 5))

ax.plot(ns, Result[0], label=r"$BigARTM$", lw = 2)
ax.set_xlabel(r'$iteration$', fontsize=18)
ax.set_ylabel(r'$perplexity$', fontsize=18)
ax.set_title('')
ax.grid(True)
```

# Experiment on NIPS collection

We did an experimnet on NIPS collection of documents. VW.LDA was run with such parameters of Dirichlet allocations $\alpha = 0.1$ and $\beta = 0.1$. BigARTM was run with only one regularizer-decorrelator with parameter $500000$.

```python
vwns = [1, 5, 10, 15, 20, 25, 30, 35, 40, 45, 50]
vwps = [2413.111, 2189.289, 2125.025, 2094.138, 2075.545, 2062.900, 2053.732, 2046.962, 2041.958, 2038.143, 2035.219]
```

## Perplexity of VW.LDA and BigARTM

```python
fig, ax = plt.subplots(figsize = (8, 5))

ax.plot(ns, Result[0], label = r"$BigARTM$", lw = 2)
ax.plot(vwns, vwps, label = r"$VW.LDA$", lw = 2)
ax.legend(loc=1) # upper left corner
ax.set_xlabel(r'$iteration$', fontsize=18)
ax.set_ylabel(r'$perplexity$', fontsize=18)
ax.set_title('');
ax.grid(True)
```

This graph shows that BigARTM topic model with only one regularizer decorellator shows better perplexity than VowpalWabbit.LDA topic model. If we find the right path of regularization, we will get better results than now. So, the idea of additive regularization is worth it to be researched further. Further experiments will include experiments on big collections (such as $Pubmed$) and will use more different regularizers.

## Sparsity of matrices $\Phi$ and $\Theta$ in BigARTM

```python
fig, axes = plt.subplots(1, 2, figsize=(10,5))

idx = 1
y = [r'$\Phi Spaprsity$', r'$\Theta Sparsity$']


for ax in axes:
    ax.plot(ns, Result[idx], label=r"$BigARTM$", lw = 2)
    #ax.legend(loc=1) # upper left corner
    ax.set_xlabel(r'$iteration$', fontsize=18)
    ax.set_ylabel(y[idx-1], fontsize=18)
    ax.grid(True)
    idx += 1

fig.tight_layout()
```

This graph shows that sparsity of matrix $\Phi$ is growing with the number of iterations (to 23% at 50 iterations) and sparsity of matrix $\Theta$ is about 0%. This can be explained with the fact that regularizers are not used in our experiment. If we run experiment on big collection and use $smooth$_$sparse$ $regularizers$, matrices will be sparser.
