```python
%matplotlib inline
import matplotlib.pyplot as plt

import numpy as np

import scipy
from scipy import ndimage
from skimage.morphology import remove_small_objects, skeletonize
from skimage.filters import threshold_otsu, sobel
from skimage.feature import canny, peak_local_max
from skimage.transform import hough_circle
from skimage.draw import circle_perimeter, circle
from skimage import color
from skimage.measure import label, perimeter, mesh_surface_area
```

```python
def binary_max_radius(image):
    i = 200
    for j in range(1, 9):
        image1 = ndimage.grey_dilation(image, size = (i, i))
        if (not np.any(1 - image1)):
            i = i - 200/(2**j)
        else:
            i = i + 200/(2**j)
    return i
```

```python
num=1
data = "D:/AlexGrinch/MIPT/IAD/Lab_2-2015/Service_" + str(num) + ".bmp"
im = scipy.misc.imread(data, flatten=True).astype(np.uint8)
edges = canny(im, sigma = 1)     
edges = ndimage.grey_dilation(edges, size = (2, 2))

binary = im > threshold_otsu(im)

res = (binary+edges)
res = np.bool_(res)
remove_small_objects(res, 800, in_place = True)

res = ndimage.grey_dilation(res, size = (3, 3))

res = np.bool_(1 - res)
remove_small_objects(res, 2000, in_place = True)
res = 1 - res

#res = skeletonize(res)

label_image = label(res)
#label_image = (label_image == 2)
number_of_objects = label_image.max()
for i in range(1, number_of_objects+1):
    service_object = (label_image == i)
    #service_object = ndimage.grey_erosion(service_object, size = (10, 10))
    feature1 = (binary_max_radius(1-service_object)/float(service_object.sum()))*1000
    #print i, feature1
    if (feature1 < 4.52):
        print "Nozh"
    elif(feature1 > 6):
        print "Lozhka"
    else: 
        print "Vilka"
    #print perimeter(service_object)

for i in xrange(3):
    fig, (ax1, ax2, ax3, ax4) = plt.subplots(ncols = 4, nrows=1, figsize=(15, 10))

    ax1.imshow((label_image == 4*i+1), cmap = plt.cm.gray)
    ax2.imshow((label_image == 4*i+2), cmap = plt.cm.gray)
    ax3.imshow((label_image == 4*i+3), cmap = plt.cm.gray)
    ax4.imshow((label_image == 4*i+4), cmap = plt.cm.gray)
    
fig, ax = plt.subplots(ncols=1, nrows=1, figsize=(15, 10))
ax.imshow(label_image, cmap=plt.cm.gray)
```
